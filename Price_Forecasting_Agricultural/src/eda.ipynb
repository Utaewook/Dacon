```python
df_dict = {key: group for key, group in df_raw.groupby('품목명')}
```

```python
list(df_dict.keys())
```

```python
for name, data in df_dict.items():
    print(f'\n품목명 = {name}')
    for column in ['품종명','거래단위','등급']:
        unique_values = data[column].unique()
        print(f"\t컬럼명: {column}")
        print(f"\t고유 값들: {unique_values}\n")
```

#### 각 품목에 따라 컬럼명과 그에 따른 값들을 살펴 보았을때 다음과 같은 사실을 알 수 있다.

1. 품종명이 여러 개 이며, 이에 따라 가격이 달라질 수 있다.
2. 거래 단위가 여러 개 이며, 이에 따라 가격이 달라질 수 있다.
3. 등급 또한 여러 개 이며, 이에 따라 가격이 달라질 수 있다.

품종명은 one hot encoding으로 처리, 거래 단위는 int값으로 변경, 등급 또한 int값으로 변경하며, 높은 등급일 수록 int값을 크게 인코딩해서 전처리 한다.
